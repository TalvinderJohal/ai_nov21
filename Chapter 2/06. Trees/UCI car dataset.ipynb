```python
import numpy as np
import pandas as pd
```

```python
car_data = "/Users/talvinderjohal/Desktop/Talvinder Strive Course/ai_nov21/Chapter 2/06. Trees/car.data"
pd.read_csv(car_data, header=None)
```

```python
columns = ['buying','maint','doors','persons','lug_boot','safety','class_vals']
```

```python
df = pd.read_csv(car_data, names=columns)
df
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
data = df.apply(le.fit_transform)
data
```

```python
X, y = data.iloc[:, :-1], data.iloc[:, -1]
```

```python
y
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0, test_size=0.3)
```

```python
X_train
```

with a random state of 0 and default criterion accuracy score = 97.30%
with a random state of 0 and criterion = entropy accuracy score = 97.8%
with a random state of 0 and max_depth = 5 accuracy score = 84.39%
with a random state of 0 and max_depth = 10 accuracy score = 96.72%
with a random state of 0 and max_depth = 100 accuracy score = 97.88%

```python
dt_clf = DecisionTreeClassifier(random_state=0, max_depth=100)
dt_clf = dt_clf.fit(X_train, y_train)
```

```python
y_pred = dt_clf.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
print(accuracy_score(y_test, y_pred))
```
